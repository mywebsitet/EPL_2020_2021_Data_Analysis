```python
#import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('EPL_20_21.ods',engine="odf")
```

```python
### Create Two more columns
data['MinitsPerMatch'] = (data['Mins']/data['Matches']).astype(int)
data['GoalsPerMatch'] = (data['Goals']/data['Matches']).astype(float)
data.head()
```

```python
### total Goals
total_goals = data['Goals'].sum()
print(total_goals)
```

```python
### Total Penalty Goals
total_penalty_goals = data['Penalty_Goals'].sum()
print(total_penalty_goals)
```

```python
### Total Penalty Attempted

Total_plenaty_attempted = data['Penalty_Attempted'].sum()
print(Total_plenaty_attempted)
```

```python
#pic Chart Penalty score and miss Penalty scour

plt.figure(figsize=(6,12))
plt_miss = Total_pleanty_attempted - total_penalty_goals
data1 = [plt_miss,total_penalty_goals]
label = ['Penalty Miss','Penalty Goals']
color = sns.color_palette('Set2')
plt.pie(data1,labels=label,colors=color,autopct='%.0f%%')
plt.show()
```

```python
## most Players come form nations

nations = data.groupby('Nationality').size().sort_values(ascending=False)
nations.head(10).plot(kind ='bar',figsize=(12,6),color = sns.color_palette('magma'))
```

```python
### Club in maximum players in ther squrd

data['Club'].value_counts().nlargest(5).plot(kind='bar',figsize=(12,6),color=sns.color_palette('Set2'))
```

```python
### Club in least players in ther squrd

data['Club'].value_counts().nsmallest(5).plot(kind='bar',figsize=(12,6),color=sns.color_palette('Set2'))
```

```python
### Player based on age Group

under_20 = data[data['Age']<=20]
age_20_25 = data[(data['Age']>20) & (data['Age']<=25)]
age_25_30 = data[(data['Age']>25) & (data['Age']<=30)]
above30 = data[data['Age']>30]
```

```python
X = np.array([under_20['Name'].count(),age_20_25['Name'].count(),age_25_30['Name'].count(),above30['Name'].count()])
mylabel = ["<=20",">20 & <=25",">25 & <=30","=>30"]
plt.title("Total players with Age Group" , fontsize=20)
plt.pie(X,labels=mylabel,autopct="%.1f%%")
plt.show()
```

```python
### Total under 20 players in each clube
players_under_20 = data[data['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar',figsize=(12,6),color=sns.color_palette('cubehelix'))
```

```python
## Average age player in each clube
plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club',y = 'Age',data=data)
plt.xticks(rotation = 90)
```

```python
num_players = data.groupby('Club').size()
data2 = (data.groupby('Club')['Age'].sum())/num_players
data2.sort_values(ascending=False)
```

```python
### Total assist from each club
Assist_by_club = pd.DataFrame(data.groupby('Club' , as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid", palette="pastel",color_codes=True)
ax = sns.barplot(x='Club',y='Assists',data=Assist_by_club.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel("Club",fontsize=20)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,20)
plt.title("Plot of Club Vs Total Assists")

```

```python
## Top 10 assists
top_10_assists = data[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top_10_assists
```

```python
### Total assist from each club
Goals_by_club = pd.DataFrame(data.groupby('Club' , as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", palette="pastel",color_codes=True)
ax = sns.barplot(x='Club',y='Goals',data=Goals_by_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel("Club",fontsize=20)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title("Plot of Club Vs Total Goals")
```

```python
### pic chart -goals with assists and with out assist

plt.figure(figsize=(6,12))
assist = data['Assists'].sum()
data3 = [total_goals-assist,assist]
label = ['Goals with Out Assist','Goals with Assists']
color = sns.color_palette('Set2')
plt.pie(data3,labels=label,colors=color,autopct='%.0f%%')
plt.show()
```
